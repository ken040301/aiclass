```python
import matplotlib as mpl
from matplotlib.font_manager import fontManager

fontManager.addfont("ChineseFont.ttf")
mpl.rc('font', family="ChineseFont")
```

```python
import matplotlib.pyplot as plt
student1 = {"國文":78, "英文":80, "數學":92, "歷史":75, "探索":85}
student2 = {"國文":92, "英文":60, "數學":85, "歷史":62, "探索":73}
#畫線圖, x軸為科目 ,y軸為分數
subjects = list(student1.keys())
scores1 = list(student1.values())
scores2 = list(student2.values())

figure=plt .figure(figsize=(20,4))
axes1=figure.add_subplot(1,2,1)
#圖一的點圖形為星形
axes1.plot(subjects,scores1,color="blue",marker="*",linestyle="-.")
axes2=figure.add_subplot(1,2,2)
axes2.plot(subjects,scores2,color="orange",marker="o",linestyle="--")
axes1.set_xlabel("科目")
axes2.set_xlabel("科目")
axes1.set_ylabel("分數")
axes2.set_ylabel("分數") 
#改變這兩個徒的y軸為最大值+5為最小值-5
axes1.set_ylim(min(scores1)-5,max(scores1)+5)   
axes2.set_ylim(min(scores2)-5,max(scores2)+5)
# 讓x軸的數值起點與終點間距相等於左右邊框
axes1.set_aspect(aspect='auto', adjustable='datalim')
axes2.set_aspect(aspect='auto', adjustable='datalim')
#我希望每個分數可以出現在點上面一點
for i, score in enumerate(scores1):
    axes1.text(subjects[i], score + 1, str(score), ha='center')
for i, score in enumerate(scores2):
    axes2.text(subjects[i], score + 1, str(score), ha='center')
#希望每個單位就有格線
axes1.grid(True, which='both', axis='y', linestyle='--', linewidth=0.5)
axes2.grid(True, which='both', axis='y', linestyle='--', linewidth=0.5)
#再做一個圖示把圖一圖二合併成一個新的圖3且有圖一格圖二的格式
figure3=plt .figure(figsize=(10,4))
axes3=figure3.add_subplot(1,1,1)
axes3.plot(subjects,scores1,color="blue",marker="*",linestyle="-.",label="學生1")
axes3.plot(subjects,scores2,color="orange",marker="o",linestyle="--",label="學生2")
axes3.set_xlabel("科目")   
axes3.legend()
axes3.set_ylabel("分數")
axes3.set_ylim(min(min(scores1), min(scores2))-5, max(max(scores1), max(scores2))+5)
axes3.set_aspect(aspect='auto', adjustable='datalim')
for i, score in enumerate(scores1):
    axes3.text(subjects[i], score + 1, str(score), ha='center', color="blue")
for i, score in enumerate(scores2):
    axes3.text(subjects[i], score + 1, str(score), ha='center', color="orange")
axes3.grid(True, which='both', axis='y', linestyle='--', linewidth=0.5)
plt.show()
```
