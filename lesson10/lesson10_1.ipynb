```python
try:
    money = int(input("請輸入金額:"))
    if money < 0:
        raise Exception("金額不可以小於0") 
except ValueError:
    print("格式錯誤")
except Exception as error:
    print(error)
else:
    print(money)
finally:
    print("應用程式結束")
```

```python
class Person:
    def __init__(self,name:str,age:int):
        self.name = name
        self.age =age
    def description(self):
        print(f"我的名字叫{self.name}")
        print(f"我的age是:{self.age}")
```

```python
p1 = Person("ken",24)
print(p1.name)
print(p1.age)
p1.description()
```

```python
p2 = Person(age=30,name='robert')
print(p2.name)
print(p2.age)
p2.description()
```

```python
class Student(Person):
    def __init__(self,name:str,age:int,english:int=60,chinese:int=60,math:int=60):
        super().__init__(name,age)
        self.__english = english
        self.__chinese = chinese
        self.__math = math
    @classmethod
    def nickname(cls):
        print("我是class的method")

    @staticmethod
    def heroname():
        print("我是static的method")
    @property
    def english(self):
        return self.__english
    @property
    def chinese(self):
        return self.__chinese
    @property
    def math(self):
        return self.__math
    def description(self):
        super().description()
        print(f"國文:{self.chinese}")
        print(f"英文:{self.english}")
        print(f"數學:{self.math}")
s1 = Student("mary",20,100,50,81)
print(s1.english)
s1.description()
```
