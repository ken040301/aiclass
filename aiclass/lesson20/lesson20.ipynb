```python
import pandas as pd
import numpy as np
data=pd.read_csv('tips.csv')
data.columns=['總票價', '小費', '是否吸菸', '日期', '時間', '人數']
data=data[['日期', '時間', '是否吸菸', '人數', '總票價', '小費']]
data['小費比例']=data['小費']/data['總票價']
data.sort_values(by='小費比例', ascending=False).iloc[:10]
```

```python
group=data.groupby('日期')
group[['小費']].mean()
```

```python
group1=data.groupby(['日期', '是否吸菸'])
s1=group1[['小費']].mean()
df1=s1.unstack()
df1
df2=s1.unstack(level='日期')
df2
```

```python
group2=data.groupby([ '日期','是否吸菸'])
s2=group2[['小費']].agg(['mean','sum','count'])
s3=s2.unstack(level='日期')
#讓小費和日期交換位子
s3.swaplevel(axis=1)
s4=group2[['小費','總票價']].agg([('平均', 'mean'),('總計', 'sum'),('筆數','count')])
s4
```

```python
def my_range(group):
    return group.max() - group.min()

grouped3 = data.groupby(['日期', '是否吸菸'])
print(len(grouped3))
df5=grouped3[['小費']].agg([('平均', 'mean'),('最大值和最小值差', my_range)])
df5['小費']['平均']
df5.loc['Sun']
```

```python
from tools import all_data_concat
import pandas as pd 

all_data = all_data_concat()
# 欄位`乘車日期`是數字格式,要轉成日期格式
all_data['乘車日期'] = pd.to_datetime(all_data['乘車日期'], format='%Y%m%d')

# 取出日期的年份,建立一個新的欄位`年份`
all_data['年份'] = all_data['乘車日期'].dt.year

# 取出期期的月份,建立一個新的欄位`月份`
all_data['月份'] = all_data['乘車日期'].dt.month

all_data.head()
```

```python
DF1=all_data.groupby(['車站名稱','年份', '月份'])[['進站人數','出站人數']].agg([('平均', 'mean'),('總計', 'sum'),('筆數','count')])
DF1
#取得DF1中七堵的2020 和2021的資料後再作合併
DF2=DF1.loc['七堵',2020]
DF3=DF1.loc['七堵',2021]
DF2.merge(DF3, left_index=True, right_index=True, suffixes=('_2020', '_2021'))
```

```python
def top(df,n=5, column='小費比例'):
    return df.sort_values(by=column, ascending=False).iloc[:n]
top(data,60)
data.groupby('是否吸菸').apply(top,n=10,include_groups=False)
```
