```python
import matplotlib.pyplot as plt
import numpy as np
index = np.array([3, 4, 5, 6, 7])
subject = ['apple', 'banana', 'orange', 'tomato', 'guava']
value1 = [5000, 6000, 7000, 8000, 9000]
value2 = [3000, 4000, 5000, 6000, 7000]
value3 = [4000, 5000, 6000, 7000, 8000]

figure = plt.figure()
axes = figure.add_subplot()
bin_width = 0.3
axes.bar(index, value1,width=bin_width, color='red')
axes.bar(index+bin_width, value2, width=bin_width, color='blue')
axes.bar(index-bin_width, value3,width=bin_width, color='green')
axes.set_xticks(index,labels=subject)
print(axes.get_xticks())
xmin,xmax,ymin,ymax=axes.axis()
xmin=xmin-0.2
xmax=xmax+0.2
axes.axis((xmin, xmax, ymin, ymax))
axes.legend(['2018年','2019年','2020年'])
#每個bin上顯示數值
for i in range(len(index)):
    axes.text(index[i], value1[i]+100, str(value1[i]), ha='center', va='bottom', fontsize=8, color='black')
    axes.text(index[i]+bin_width, value2[i]+100, str(value2[i]), ha='center', va='bottom', fontsize=8, color='black')
    axes.text(index[i]-bin_width, value3[i]+100, str(value3[i]), ha='center', va='bottom', fontsize=8, color='black')
plt.show()
```

```python
# 1家分站30天的營收

oneStore = np.random.randint(100, 10000, size=(30,))
oneStore
```

```python
# 累計每天這家店的營收
import matplotlib.pyplot as plt

oneStoreCum = oneStore.cumsum()
oneStoreCum

figure = plt.figure(figsize=(10, 5))
axes = figure.add_subplot(1,1,1)
axes.plot(range(1, 31), oneStoreCum, 'r-')
axes.set_title("one Store shop cumsum")
axes.set_xlabel("day")
axes.set_ylabel("money")
plt.show()
```

```python
# 5家店30天的營收

fiveStore = np.random.randint(100, 10000, size=(5, 30))
fiveStore
```

```python
# 累計每天5家店的營收
fiveStoreCum = fiveStore.cumsum(axis=1)
fiveStoreCum
```

```python
# 累計每天這家店的營收
import matplotlib.pyplot as plt

figure = plt.figure(figsize=(10, 5))
axes = figure.add_subplot(1,1,1)
axes.plot(range(1, 31), fiveStoreCum[0], 'r-')
axes.plot(range(1, 31), fiveStoreCum[1], 'y-')
axes.plot(range(1, 31), fiveStoreCum[2], 'g-')
axes.plot(range(1, 31), fiveStoreCum[3], 'm-')
axes.plot(range(1, 31), fiveStoreCum[4], 'c-')
axes.set_title("one Store shop cumsum")
axes.set_xlabel("day")
axes.set_ylabel("money")
plt.show()
```
