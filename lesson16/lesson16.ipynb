```python
import pandas as pd
df=pd.read_csv('各鄉鎮市區人口密度.csv')
df.columns=df.iloc[0].tolist()
df=df.drop(index=0).reset_index(drop=True)
#把最後幾行補充刪除
df= df.drop(index=df.index[-5:])
#年底人口數 和人口密度最底下有...想把它換成0
df['年底人口數']=df['年底人口數'].replace('… ', '0').astype(int)
df['人口密度']=df['人口密度'].replace('… ', '0').astype(int)
df['土地面積']=df['土地面積'].astype(float)
df
```

```python
import sys, subprocess
import importlib

# 如何將df2存成新的csv檔案
df.to_csv('各鄉鎮市區人口密度_清理後.csv', index=False)
# 如何將df2存成新的excel檔案
df.to_excel('各鄉鎮市區人口密度_清理後.xlsx', index=False)
# 確保 openpyxl 已安裝（若未安裝則透過 pip 安裝）
try:
    import openpyxl  # type: ignore
except ModuleNotFoundError:
    print("openpyxl not found. Installing...")
```

```python
import pandas as pd

# 建立一個 Series（產品銷售額）
sales = pd.Series([300, 600, 450, 800, 200, 700], 
                  index=['A', 'B', 'C', 'D', 'E', 'F'])
# 建立布林遮罩：銷售額 > 500
mask = sales > 500
print("布林遮罩：")
print(mask)
# 輸出：
# A    False
# B     True
# C    False
# D     True
# E    False
# F     True
# dtype: bool

# 使用遮罩篩選高銷售產品
high_sales = sales[mask]
print("\n銷售額 > 500 的產品：")
print(high_sales)
```
